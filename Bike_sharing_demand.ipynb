## Skewness and kurtois

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

graph = sns.histplot(df['count'], color='r', label='Skewness: {:.2f}'.format(df['count'].skew()), ax=ax)

graph.legend(loc='best')

print('Skewness : %f' % df['count'].skew())
print('Kurtosis : %f' % df['count'].kurt())
```

We can say that the data is normalized when the skew or kurtosis is -2 ~ +2

However, here, we can see that there are too many values accumulated in 0. Therefore I decided to normalize the data using log scaling.

```python
df['normalized'] = df['count'].map(lambda i: np.log(i) if i > 1 else 0)
```

```python
# Create a figure and axis using subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Use histplot (or displot) instead of deprecated distplot
graph = sns.histplot(df['normalized'], color='g', label='Skewness: {:.2f}'.format(df['normalized'].skew()), ax=ax)

# Add legend
graph.legend(loc='best')

# Display skewness and kurtosis values
print('Skewness: {:.6f}'.format(df['normalized'].skew()))
print('Kurtosis: {:.6f}'.format(df['normalized'].kurt()))

# Show the plot
plt.show()
```

we can see that the data is well distributed.

# Modeling

Gradient boosting

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df.drop('weekend', axis=1, inplace = True)
```

```python
x_train = df.drop('normalized', axis = 1).values
target_label = df['normalized'].values
x_test = df_test.values


x_train, x_val, y_train, y_val = train_test_split(x_train, target_label, test_size = 0.25, random_state = 2000)
```

Since it is not array, we convert the data from dataframe to array shape

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
Regressor = GradientBoostingRegressor(n_estimators=2000, learning_rate = 0.05, min_samples_leaf = 15
                                    , min_samples_split = 10
                                    , random_state = 4)
```

```python
Regressor.fit(x_train, y_train)
```

```python
score_train = Regressor.score(x_train, y_train)
score_val = Regressor.score(x_val, y_val)

print("train score: %f" %score_train)
print("validation score: %f" %score_val)
```
